```python
def seconds_to_hms(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    return hours, minutes, seconds
```

```python
# how to list only burnett's - create dataframe with burnett's rows
substring = 'Burnett'  # careful you have burnett and burnett's .... cleanse before :-)
sr = df[df['Item Description'].str.contains(substring, case=False)]
sr
```

```python
brands = sr['Item Description'].value_counts().reset_index()
brands
```

```python
def remove_first_name(name):
   a = name
   return a.split(' ', 1)[1]

# Create a column with Burnett's product without burnett's name
sr1 = sr.copy()
sr1['brand name'] = sr['Item Description'].copy().apply(remove_first_name)
```
