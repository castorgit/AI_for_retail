##### **Apriori Algorithm on UCI online Dataset**

see https://medium.com/@jihargifari/how-to-perform-market-basket-analysis-in-python-bd00b745b106

Using mixtend package

Using UCI online dataset (2 years)

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data0 = pd.read_csv('./data/data_online_reail_UCI_dataset_2010.csv')
data1 = pd.read_csv('./data/data_online_reail_UCI_dataset_2009.csv')
data = pd.concat([data1, data0], ignore_index=True)
```

```python
data
```

```python
data=data.dropna()
data.info()
```

```python
data_plus= data[data['Quantity']>=0]
data_plus.info()
```

```python
basket_plus = (data_plus[data_plus['Country']=='United Kingdom'].groupby(['InvoiceNo','Description'])['Quantity'] \
               .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_plus
```

```python
def encode_units(x):
    if x <=0:
        return 0
    if x>=1:
        return 1
basket_encode_plus = basket_plus.map(encode_units)
basket_encode_plus.info()
```

```python
basket_filter_plus = basket_encode_plus[(basket_encode_plus>0).sum(axis=1)>=2]
basket_filter_plus.info()
```

```python
frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.023,
                                use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))
frequent_itemsets_plus
```

```python
association_rules(frequent_itemsets_plus, metric='lift',
                 min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
